```python
#Handling missing values 
#There's no reliable way to infer a missing director,cast, rating and country from other columns (mapping)
#They don't have any -1 to 1 relationship (correlation)

df['director'] = df['director'].fillna('Not Specified')
df['cast'] = df['cast'].fillna('Not Specified')
df['country'] = df['country'].fillna('Not Specified')
df['rating'] = df['rating'].fillna('Not Rated')
```

```python
df.sample(20)
```

```python
#Filling a date with a fake placeholder date would be misleading 
#Since it's only 10 rows out of ~8,800, the safest option is usually to drop them
df = df.dropna(subset=['date_added'])
```

```python
df.shape
```

```python
#Rows where duration is missing actually have their real duration sitting in 'rating'
#Checked and verified via the Excel sheet

swapped_rows = df['duration'].isnull()

df.loc[swapped_rows, 'duration'] = df.loc[swapped_rows, 'rating']
df.loc[swapped_rows, 'rating'] = 'Not Rated'  # since the real rating for these rows is genuinely unknown

print(df[swapped_rows][['rating', 'duration']])  # confirm the fix
print(df['duration'].isnull().sum())  # should be 0
```

```python
#Clean up the whitespace and converts the column from text into a real date type (Python readable)
df['date_added'] = pd.to_datetime(df['date_added'].str.strip())
print(df['date_added'].dtype)
```

```python
#We take every row, breaks the country list apart at the commas, and creates a separate row per country 
#Instead of having one messy row
df_countries_split = df.assign(country=df['country'].str.split(', ')).explode('country')
```

```python
#Check rating values look valid (no junk/typos)
print(df['rating'].unique())
```

```python
#Check that date_added is never before release_year
#date_added- The actual date the movie/show was added to Netflix
#release_year- The year the movie/show was actually released
invalid_dates = df[df['date_added'].dt.year < df['release_year']]
print(invalid_dates.shape[0])
print(invalid_dates[['title', 'release_year', 'date_added']])
```

```python
#Drop the above records because it doesn't make sense 
#A movie/show can't be added to Netflix before release
df = df[~df.index.isin(invalid_dates.index)]
print(df.shape)
```

```python
#Split the movies and shows into separate datasets
df_movies = df[df['type'] == 'Movie'].copy()
df_tv = df[df['type'] == 'TV Show'].copy()

print("Movies:", df_movies.shape)
print("TV Shows:", df_tv.shape)

#Ensure a proper split in their individual numeric columns
df_movies['DurationMinutes'] = df_movies['duration'].str.replace(' min', '', regex=False).astype(int)

df_tv['Seasons'] = df_tv['duration'].str.extract(r'(\d+)').astype(int)

#Quick check that it worked
print(df_movies[['title', 'duration', 'DurationMinutes']].sample(5))
print(df_tv[['title', 'duration', 'Seasons']].sample(5))

#Export both datasets as separate cleaned CSVs
df_movies.to_csv('Cleaned_Netflix_Movies.csv', index=False)
df_tv.to_csv('Cleaned_Netflix_TVShows.csv', index=False)

#Also export the combined df (all titles, Movies + TV Shows together) as a third file 
#Some EDA questions (like "Movies vs TV Shows distribution") need both types in one dataset
df.to_csv('Cleaned_Netflix_All.csv', index=False)
```

```python
combo = df.assign(country=df['country'].str.split(', ')).explode('country')
combo['country'] = combo['country'].str.strip()
combo = combo[(combo['country'] != 'Not Specified') & (combo['country'] != '')]

combo = combo.assign(listed_in=combo['listed_in'].str.split(', ')).explode('listed_in')

country_genre_table = combo.groupby(['country', 'listed_in'])['show_id'].nunique().reset_index()
country_genre_table.columns = ['Country', 'Genre', 'TitleCount']

country_dd2 = widgets.Dropdown(options=['All Countries'] + sorted(country_genre_table['Country'].unique()), value='All Countries', description='Country:')
genre_dd2 = widgets.Dropdown(options=['All Genres'] + sorted(country_genre_table['Genre'].unique()), value='All Genres', description='Genre:')
output_cg = widgets.Output()

def update_cg_table(change=None):
    output_cg.clear_output()
    with output_cg:
        filtered = country_genre_table.copy()
        if country_dd2.value != 'All Countries':
            filtered = filtered[filtered['Country'] == country_dd2.value]
        if genre_dd2.value != 'All Genres':
            filtered = filtered[filtered['Genre'] == genre_dd2.value]
        display(filtered.sort_values('TitleCount', ascending=False).reset_index(drop=True))

country_dd2.observe(update_cg_table, names='value')
genre_dd2.observe(update_cg_table, names='value')
display(country_dd2, genre_dd2, output_cg)
update_cg_table()
```

```python
#Country-level average duration/seasons
#Does content from certain countries tend to run longer?

movies_country_split = df_movies.assign(country=df_movies['country'].str.split(', ')).explode('country')
movies_country_split['country'] = movies_country_split['country'].str.strip()
movies_country_split = movies_country_split[movies_country_split['country'] != 'Not Specified']

avg_duration_country = (
    movies_country_split.groupby('country')['DurationMinutes'].mean()
    .sort_values(ascending=False).head(100).reset_index()
)

tv_country_split = df_tv.assign(country=df_tv['country'].str.split(', ')).explode('country')
tv_country_split['country'] = tv_country_split['country'].str.strip()
tv_country_split = tv_country_split[tv_country_split['country'] != 'Not Specified']

avg_seasons_country = (
    tv_country_split.groupby('country')['Seasons'].mean()
    .sort_values(ascending=False).head(100).reset_index()
)

fig = make_subplots(rows=1, cols=2, subplot_titles=('Avg. Movie Duration by Country (Top 100)', 'Avg. Seasons by Country (Top 100)'),
                     horizontal_spacing=0.25)

fig.add_trace(go.Bar(x=avg_duration_country['DurationMinutes'], y=avg_duration_country['country'],
                       orientation='h', marker_color='#F4A261'), row=1, col=1)
fig.add_trace(go.Bar(x=avg_seasons_country['Seasons'], y=avg_seasons_country['country'],
                       orientation='h', marker_color='#2A9D8F'), row=1, col=2)

fig.update_yaxes(categoryorder='total ascending', row=1, col=1)
fig.update_yaxes(categoryorder='total ascending', row=1, col=2)
fig.update_layout(height=650, showlegend=False, title_text='Average Duration/Seasons by Country')
fig.show()
```
